## 가해 운전자 연령과 피해 운전자 연령 시각화

```python
offend_filled["가해운전자 연령"].unique()
```

```python
offend_filled['피해운전자 연령'].value_counts()
```

```python
offend_filled['가해운전자 연령'] = offend_filled['가해운전자 연령'].replace('미분류', 0)
offend_filled['피해운전자 연령'] = offend_filled['피해운전자 연령'].replace('미분류', 0)
```

```python
offend_filled['피해운전자 연령'].unique()
```

```python
import re

# 숫자 부분만 추출하는 함수 정의
def remove_non_numeric(age):
    if isinstance(age, str):
        # 정규 표현식을 사용하여 숫자만 추출
        numeric_part = re.findall(r'\d+', age)
        if numeric_part:
            return int(numeric_part[0])
    return age

# '피해운전자 연령' 컬럼에 함수 적용
offend_filled['피해운전자 연령'] = offend_filled['피해운전자 연령'].apply(remove_non_numeric)
offend_filled['가해운전자 연령'] = offend_filled['가해운전자 연령'].apply(remove_non_numeric)


# 결과 확인
print(offend_filled['피해운전자 연령'].unique())
print(offend_filled['가해운전자 연령'].unique())
```

```python
import pandas as pd

# 라벨링 함수
def label_age(age):
    if age == 0:
        return '0'
    elif 1 <= age <10:
        return '10대 미만'
    elif 10 <= age <20:
        return '10대'
    elif 20 <= age <30:
        return '20대'
    elif 30 <= age <40:
        return '30대'
    elif 40 <= age <50:
        return '40대'
    elif 50 <= age <60:
        return '50대'
    elif 60 <= age <70:
        return '60대'
    elif 70 <= age <80:
        return '70대'
    else:
        return '80세 이상'

# '가해운전자 연령' 컬럼에 라벨링 적용
offend_filled['가해운전자 연령대'] = offend_filled['가해운전자 연령'].apply(label_age)
age_groups = ['0','10대 미만', '10대', '20대', '30대', '40대','50대','60대','70대','80세 이상']
offend_age_group_counts = offend_filled['가해운전자 연령대'].value_counts().reindex(age_groups, fill_value=0)

# '피해운전자 연령' 컬럼에 라벨링 적용
offend_filled['피해운전자 연령대'] = offend_filled['피해운전자 연령'].apply(label_age)
victim_age_group_counts = offend_filled['피해운전자 연령대'].value_counts().reindex(age_groups, fill_value=0)


print(offend_age_group_counts)
print(victim_age_group_counts)
```

```python
# 가해자 비중 원그래프 생성
plt.figure(figsize=(10, 7))
plt.pie(offend_age_group_counts, labels=age_groups, autopct='%1.1f%%', startangle=140,  labeldistance=1.1, pctdistance=0.75, colors=sns.color_palette("pastel"))
plt.legend( age_groups, title="연령대", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.title('가해운전자 연령대 분포')
plt.axis('equal')  # 원을 원형으로 유지
plt.show()
```

```python
# 피해자 비중 원그래프 생성
plt.figure(figsize=(10, 7))
plt.pie(victim_age_group_counts, labels=age_groups, autopct='%1.1f%%', startangle=140,  labeldistance=1.1, pctdistance=0.75, colors=sns.color_palette("pastel"))
plt.legend( age_groups, title="연령대", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.title('피해운전자 연령대 분포')
plt.axis('equal')  # 원을 원형으로 유지
plt.show()
```
